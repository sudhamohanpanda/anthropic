```python
# I generated this code using claude.
import textstat

def grade_readability(response_text):
    # Flesch Reading Ease (higher = easier to read)
    flesch_score = textstat.flesch_reading_ease(response_text)
    
    # Flesch-Kincaid Grade Level
    grade_level = textstat.flesch_kincaid_grade(response_text)
    
    # Gunning Fog Index
    fog_index = textstat.gunning_fog(response_text)
    
    return {
        "flesch_reading_ease": flesch_score,
        "grade_level": grade_level,
        "gunning_fog": fog_index
    }

# Example: assert response is readable at or below 8th grade level
def eval_pass(response_text, max_grade=8):
    score = textstat.flesch_kincaid_grade(response_text)
    return score <= max_grade
```
